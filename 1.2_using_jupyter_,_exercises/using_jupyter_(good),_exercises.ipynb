```python
%timeit squares = [i**2 for i in range(10)]
```

```python
%%timeit squaress = list()
for i in range(10):
    squaress.append(i**2)
```

The numbers will mean that the second code is more concised and hence, needs less time and power from the computer to run the code, though both codes gives the same output
